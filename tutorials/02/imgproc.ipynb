```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
w, h = width, height = 1280, 720
```

```python
data = np.fromfile('frame20.yuv', dtype=np.uint8)
```

```python
data.shape
```

```python
start, end = 0, w*h
yplane = data[:w*h].reshape(h,w)
plt.imshow(yplane)
plt.show()
```

```python
start, end = end, end+w//2*h//2
start, end
```

```python
uplane = data[start:end].reshape((h//2,w//2))
plt.imshow(uplane)
plt.show()
```

```python
start, end = end, end+w//2*h//2
start, end
```

```python
vplane = data[start:end].reshape((h//2,w//2))
plt.imshow(vplane)
plt.show()
```

```python
yuv = data.reshape((-1,1280))
yuv.shape
```

```python
bgr = cv2.cvtColor(yuv, cv2.COLOR_YUV2BGR_I420)
plt.imshow(bgr)
```

```python
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```
