```python
udata = np.fromfile('1', dtype=np.uint8)
udata.shape
```

```python
uimg = udata.reshape((360, 640))
plt.imshow(uimg)
plt.show()
```

```python
vdata = np.fromfile('2', dtype=np.uint8)
vdata.shape
```

```python
vimg = vdata.reshape((360, 640))
plt.imshow(vimg)
plt.show()
```

```python
type(ydata)
```

```python
yuv_data = np.concatenate((ydata, udata, vdata))
```

```python
yuv_data.shape
```

```python
w, h = width, height = 1280, 720
```

```python
yuv0 = yuv_data.reshape((height*3//2, width))
yuv0.shape
```

```python
bgr0 = cv2.cvtColor(yuv0, cv2.COLOR_YUV2BGR_I420)
plt.imshow(bgr0)
```

```python
rgb0 = cv2.cvtColor(bgr0, cv2.COLOR_BGR2RGB)
plt.imshow(rgb0)
```

```python
fp = open("sample.yuv","wb")
yuv_data.tofile(fp);
fp.close()
```

### display YUV frame

```python
frame_size = width * height * 3//2
```

```python
f = open('out.yuv', 'rb')
```

```python
yuv = np.frombuffer(f.read(frame_size), dtype=np.uint8).reshape((height*3//2, width))
f.close()
```
